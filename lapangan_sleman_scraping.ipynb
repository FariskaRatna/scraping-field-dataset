```python
def convert_to_float(value):
    if isinstance(value, str):
        return float(value.replace(',', '.'))
    else:
        return value

sleman_df['luas_wilayah_(km2)'] = sleman_df['luas_wilayah_(km2)'].apply(convert_to_float)
```

```python
def radius(luas):
    return np.sqrt((luas*1000000/np.pi))
```

```python
map_client = googlemaps.Client("AIzaSyD4NxoIzL7sZcUB-ZMrGXerlzedHm1YcZA")
```

```python
location = pd.DataFrame()
```

```python
def place_location(df, nama_kecamatan, nama_kabupaten):
    location_name = f"{nama_kecamatan} {nama_kabupaten}"
    response = map_client.places(query = location_name)
    results = response.get('results')
    if results:
      print(results[0]['geometry']['location'])
      df = df.append(results[0]['geometry']['location'], ignore_index = True)
    else:
      print("No results found for query: ", location_name)

    return df
```

```python
for index, row in sleman_df.iterrows():
    location = place_location(location, row['kecamatan'], 'Kabupaten Sleman Yogyakarta')
```

```python
sleman_df["lat"] = location["lat"]
sleman_df["lng"] = location["lng"]
```

```python
place_list = pd.DataFrame()
```

```python
def search_places (location, search_string, distance, place_list):
    response = map_client.places_nearby(
        location = location,
        keyword = search_string,
        radius = distance,
    )
    results = response.get('results')
    place_list = place_list.append(results)

    return place_list
```

```python
import re
def search_word(text, word, type):
    if (type == "address"):
        words = text.split(sep=', ')  # Split the text into individual words
    elif (type == "name"):
        words = re.split(r"[', ']", text)

    count = 0
    value = False
    # print(words)

    for w in words:
        if w == word:
            count+=1
            value = True  # Word found
    return value, count # Word not found
```
